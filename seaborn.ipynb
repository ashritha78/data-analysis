```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
print(tips)
sns.scatterplot(x="total_bill",y="tip",data=tips,label='tips')
plt.title("Scatter Plot of Total Bill vs Tip")
plt.xlabel("Total Bills($)")
plt.ylabel("Tip($)")
plt.legend()
plt.show()
```

```python
iris=sns.load_dataset('iris')
print(iris)
sns.scatterplot(x="sepal_length",y="sepal_width",data=iris,label='iris')
plt.title("Scatter plot of iris")
plt.xlabel("Sepal Length($)")
plt.ylabel("Sepal width($)")
plt.legend()
plt.show()
```

```python
iris=sns.load_dataset('iris')
print(iris)
sns.barplot(x="sepal_length",y="sepal_width",data=iris,label='iris')
plt.title("Scatter plot of iris")
plt.xlabel("Sepal Length($)")
plt.ylabel("Sepal width($)")
plt.legend()
plt.show()
```

```python
tips=sns.load_dataset('tips')
print(tips)
sns.boxplot(x="total_bill",y="tip",data=tips)
plt.title("Scatter plot of tips")
plt.xlabel("Total Bills($)")
plt.ylabel("Tip($)")
plt.show()
```

```python
tips=sns.load_dataset('tips')
print(tips)
sns.violinplot(x="total_bill",y="tip",data=tips)
plt.title("Scatter plot of tips")
plt.xlabel("Total Bills($)")
plt.ylabel("Tip($)")
plt.show()
```

```python
iris=sns.load_dataset("iris")
correlation_matrix=iris.corr()
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm")
plt.title("correlation Heatmap of iris Dataset")
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#Load example dataset
tips=sns.load_dataset("tips")
print(tips)
#create a scatter plot
sns.countplot(x="day",data=tips,label='tips')
plt.legend(loc='best')
plt.title("scatter plot of Day of the week vs. Total bill")
plt.xlabel("Day of the week($)")
plt.ylabel("Total bill($)")
plt.show()
```

```python
tips=sns.load_dataset('tips')
print(tips)
sns.lmplot(x="total_bill",y="tip",data=tips)
plt.title("Scatter plot of tips")
plt.xlabel("Total Bills($)")
plt.ylabel("Tip($)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic=sns.load_dataset("titanic")
g=sns.FacetGrid(titanic,col="class")
g.map(sns.histplot,"age")
plt.subplots_adjust(top=0.8)
g.fig.suptitle("Age Distribution by Passenger Class")
plt.show()
```
